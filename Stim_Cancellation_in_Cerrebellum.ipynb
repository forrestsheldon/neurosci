## Motivation

The aquatic environment presents a challenge for sensing.  First, it is quite dark, so visual information is of limited use.  Second, it requires very active navigation and so most animals must use their appendages primarily for locomotion.  On the up side, moving in three dimensions means there's usually nothing around you anyways and so appendages for grabbing and feeling are probably of limited value anyway (I guess that's why fish don't have arms).

Fish need to eat though and so need some way of navigating this dark environment and finding prey.  In response, evolution has granted them the ability to generate and sense the electric field in their vicinity.  Via a lateral sensing organ, they can 'see' by sensing the electric field distortions around them that may be due to objects or prey.

Sensory information is processed by the cerebellum, a globular structure possessed by all vertebrates (and some invertebrates I believe) whose architecture has proven difficult to analyze.  In the (approximate) words of Chuck Stevens, these regions have resisted being assigned a map to the extent that it now seems that the extreme opposite must be true; the location of the cells tells you *nothing* about the information being processed. Several of these regions have a particular structure that seems to be enacting a random linear transformation.  Such a transformation is information preserving if it possesses a particular property (Restricted-Isometry Property) and this would also explain the desirability of the 'anti-map' phenomenon.

In order to understand how the cerrebellum is processing information, it is desireable to work with a very simple sensory system that (ideally) know something about.

*In step the weakly electric fish*

In a series of experiments, it has been demonstrated that the electric fish cerrebellum acts to cancel predictable sensory signals that correlate with bodily motor functions. (Read, the fishes brain cancels out the part of the signal that corresponds to it swimming in order to see the prey.)  If we can come up with a model that explains how the cerrebellum is doing this, we can put together some Hodgkin-Huxley neurons and test it out, and maybe even throw it on a chip and make run a little robot! Neato! Plus it might extend to more complicated cerrebellar functioning.

## Cerebellar Anatomy in Weakly Electric Fish

#### From: Cerebellum-like structures

In addition to the cerebellum, there are several areas that resemble it strongly which are termed *cerebellum-like structures*.

Across species, the cerebellum has two inputs:

* Mossy fibers carry sensory and motorinformation to Granule cells which send out parallel fibers to the arbors of Purkinje cells
* Each Purkinje cell also recieves a very strong inhibitory coupling from climbing fibers arising in the inferior olive

While in cerebellum-like structures:

* Mossy fiber-granule cell-parallel fiber structures send information to the arbors of Purkinje cells
* Direct sensory information also synapses on the *basilar* dendrites of the Purkinje cells (also called principal cells)



#### The Following is from: From sparks to spikes: information processing in the electrosensory systems of fish

Overview of Electrosensory system: Electroreceptors possess cells that synapse on primary afferent fibers.  Fish can passively sense external fields and actively sense resistances and capacitances of nearly objects (electric color!) though EODs from electric organs. Electroreceptors that are sensitive to their own EODs are called tuberous receptors and divide into amplitude and timing sensitive receptors.  The pathways for these two types remain separate at the inital stages of processing, perhaps similarly to the auditory systems of vertebrates.  This work focuses on the amplitude coding pathway.  Afferent fibers terminate deep in the ELL, and their input is relayed to principal cells (sometimes through interneurons) that project out of the ELL.  Principal cells (pyramidal cells in gymnotiforms and efferent cells in mormyrids) come in two varieties: E-cells - excited by an increase in EOD amplitude and I-cells - inhibited by an increase in EOD amplitude.  Mormyrids also have Purkinje like cells providing inhibitory inputs to efferent cells and each other.  Principal and Purkinje cell apical dendrites extend into a molecular layer and recieve inputs from granule cells that carry information about proprioception, motor commands and higher processing.
