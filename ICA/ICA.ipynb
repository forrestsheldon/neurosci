# Blind Source Separation and ICA

**Blind Source Separation (BSS)** is the problem of inferring multiple source signals by recording multiple signal mixtures without knowledge of the initial sources. In it's most general form,

given a set of signal mixtures $x$ such that,
$$x = F(s)$$
extract the source signal $s$.

Unmixing typically proceeds along some subset of three paths:

* **Independence** Sources are created by unrelated physical processes and should thus be statistically independent.  Signal mixtures, as they are created by simple physical means of combination typically are not.  Thus by extracting independent signals from a signal mixture, we can assume that they come from some underlying sources
* **Normality** Source signals are created by a variety of physical processes that many times are non-gaussian when binned.  By the central limit theorem, signal mixtures will look more gaussian than the signal mixtures.
* **Complexity**  The temporal complexity of a signal mixture is greater than or equal to the complexity of the simplest source signal.  Thus, by extracting the signal of minimal complexity, we may assume that it is a source signal.

Let's observe this in a model system.

#### Importing Data

```python
from scikits.audiolab import wavread, play
import numpy as np
import matplotlib.pyplot as plt
```

We import two wav files that we will mix.  One of music and one of some stock market news.

```python
Music_dat, Music_fs, Music_oth = wavread("Music_source.wav")
play(Music_dat, Music_fs)
```

```python
Stock_dat, Stock_fs, Stock_oth = wavread("Stock_source.wav")
play(Stock_dat, Stock_fs)
```

#### Normality

Binning each signal separately we see that they are not quite gaussian (the speech signal far moreso).

```python
bars, bins, patches = plt.hist(Music_dat, bins=50)
```

```python
bars, bins, patches = plt.hist(Stock_dat, bins=50)
```

Now we mix the two signals and find that the mixture appears somewhat more gaussian than either of the original signals.

```python
mix_mat = 2 * np.random.random((2,2))- 1
```

```python
mix1 =  mix_mat[0, 0] * Music_dat + mix_mat[0, 1] * Stock_dat
mix2 =  mix_mat[1, 0] * Music_dat + mix_mat[1, 1] * Stock_dat
```

```python
bars, bins, patches = plt.hist(mix1, bins=50)
```

#### Independence

Plotting the signals on orthogonal axes we see that the signals appear uncorrelated, as we would expect for independent signals.

```python
fig1, ax1 = plt.subplots()
ax1.scatter(Music_dat, Stock_dat, marker='.', alpha=0.05)
```

When we plot the signal mixtures this is no longer the case.  The signals appear correlated and the point cloud becomes skewed. This will vary somewhat run to run

```python
fig2, ax2 = plt.subplots()
ax2.scatter(mix1, mix2, marker='.', alpha=0.05)
```
